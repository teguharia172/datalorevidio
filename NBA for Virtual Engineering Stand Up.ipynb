## Playoff

```python
df_play = pd.read_csv("Schedule Performance Play Off NBA_2023-04-26 02_50_21 PM.csv", index_col=0, parse_dates=True)
df_play.head()
```

```python
df_play.plot(
    kind="scatter",
    x="total_duration",
    y="total_plays",
    s=35,
    figsize=(12, 8),
    fontsize = 12
)

plt.title("NBA Playoff Performance\n(2023/04/15 - 2023/04/21)", fontsize=17)

plt.xlabel("Total Duration (Hour)", fontsize=12)
plt.ylabel("Total Plays", fontsize=12)

plt.xlim(xmin=1_000)
plt.ylim(ymin=0)

plt.axhline(y=df_play["total_plays"].median(), color="#C70039", lw=2, ls="--")

plt.annotate(
    text=f"Sacramento Kings vs Golden State Warriors\n(2023/04/21)",
    xy=(df_play["total_duration"].iloc[0] - 500, df_play["total_plays"].iloc[0]),
    ha="right",
    va="center",
    fontsize=10
)

plt.annotate(
    text="Chicago Bulls vs Miami Heat\n(2023/04/15)",
    xy=(df_play["total_duration"].iloc[-1] + 500, df_play["total_plays"].iloc[-1]),
    ha="left",
    va="center",
    fontsize=10
)

plt.annotate(
    text="Atlanta Hawks vs Boston Celtics\n(2023/04/19)",
    xy=(df_play["total_duration"].iloc[-2] - 500, df_play["total_plays"].iloc[-2]),
    ha="right",
    va="center",
    fontsize=10
)

plt.annotate(
    text="Median Total Plays",
    xy=(25_000, 41_000),
    ha="center",
    c="#C70039",
    fontsize = 12
)

plt.grid(axis="x")
plt.savefig("nba_playoff_performance.png", dpi=400)
plt.show()
```

## Running Sum

```python
df_sum = pd.read_csv("Running Sum Total Plays NBA Regular Season _2023-04-26 02_42_34 PM.csv", index_col=0, parse_dates=True)
df_sum.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Left axis
plot1 = ax.plot(df_sum.index, df_sum.running_sum_total_plays, lw=1.5, label="Running Sum Total Plays")

ax.set_xlim(xmax=np.max(df_sum.index) + pd.Timedelta(days=21))
ax.set_ylim(ymin=0)
ax.set_yticks(ax.get_yticks(), [numerize.numerize(i) for i in ax.get_yticks()])
ax.set_ylabel("Running Sum Total Plays", fontsize=12, labelpad=15)
ax.grid(axis="x")

for label in ax.get_yticklabels():
    label.set_color("#008FD5")

ax.annotate(
    text=numerize.numerize(int(df_sum["running_sum_total_plays"].iloc[-1]), 1),
    xy=(
        df_sum.tail(1).index + pd.Timedelta(days=2),
        df_sum["running_sum_total_plays"].iloc[-1]
    ),
    ha="left",
    va="center",
    fontsize=11
)
    
# Right axis
ax2 = ax.twinx()
plot2 = ax2.plot(df_sum.index, df_sum.running_sum_total_games, lw=1.5, c="#C70039", label="Running Sum Total Games")

ax2.set_ylabel("Running Sum Total Games", fontsize=12, labelpad=15)
ax2.grid(False)

for label in ax2.get_yticklabels():
    label.set_color("#C70039")
    
ax2.annotate(
    text=f'{df_sum["running_sum_total_games"].iloc[-1]} Games',
    xy=(
        df_sum.tail(1).index + pd.Timedelta(days=2),
        df_sum["running_sum_total_games"].iloc[-1]
    ),
    ha="left",
    va="center",
    fontsize=11
)   

# Overall figure
plots = plot1 + plot2
labels = [plot.get_label() for plot in plots]
ax.legend(plots, labels, loc="upper left", fontsize=10)
ax.set_title("Cumulative NBA Regular Season Performance\n(2022/10/20 - 2023/04/09)", fontsize=15, )

plt.savefig("cumulative_nba_regular_season_performance.png", dpi=200)
plt.show()
```
